```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Olympic_Games_host_cities'
```

```python
tables = pd.read_html(url)
tables
```

```python
type(tables)
```

```python
venues_df = tables[2]                 
venues_df.head(10)
```

```python
del venues_df['City']
del venues_df['Summer']
del venues_df['Winter']
del venues_df['Ref']
venues_df
```

```python
venues = venues_df.rename(columns={'City.1':'city','Country':'country', 'Year':'year','Continent':'continent','Opening ceremony':'opening_ceremony','Closing ceremony':'closing_ceremony'})
venues.head()
```

```python
rds_connection_string = "postgres:Jacky0509@localhost:5432/Olympic_DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
venues.to_sql(name='venues', con=engine, if_exists='append', index=False)
```
