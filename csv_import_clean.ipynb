```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file1 = "Resources/athlete_events.csv"
full_athlete = pd.read_csv(csv_file1)
full_athlete.head()
```

```python
csv_file2 = "Resources/noc_regions.csv"
regions = pd.read_csv(csv_file2)
regions.head()
```

```python
regions = regions.rename(columns={'NOC':'noc'})
regions.head()
```

```python
teams_df = full_athlete[["ID", "Team"]].copy()
teams_df.head()
```

```python
teams_df = teams_df.rename(columns={'ID':'athlete_id', 'Team':'team'})
teams_df.head()
```

```python
nodup_teams_df = pd.DataFrame.drop_duplicates(teams_df)
nodup_teams_df.head()
```

```python
athlete_df = full_athlete[["ID", "Name", "Sex"]].copy()
athlete_df.head()
```

```python
athlete_df = athlete_df.rename(columns={'ID':'athlete_id', 'Name':'name', 'Sex':'sex'})
athlete_df.head()
```

```python
nodup_athlete_df = pd.DataFrame.drop_duplicates(athlete_df)
nodup_athlete_df.head()
```

```python
athlete_data = full_athlete[["ID", "Year", "Sex", "Age", "Height", "Weight", "Team"]].copy()
athlete_data.head()
```

```python
athlete_data = athlete_data.rename(columns={'ID':'athlete_id', 'Year': 'year', 'Sex':'sex','Age':'age', 'Height':'height', 'Weight':'weight', 'Team': 'team'})
athlete_data.head()
```

```python
nodup_athlete_data = pd.DataFrame.drop_duplicates(athlete_data)
nodup_athlete_data.head()
```

```python
sport_df = full_athlete[["ID", "Sport", "Event"]].copy()
sport_df.head()
```

```python
sport_df = sport_df.rename(columns={'ID':'athlete_id', 'Sport':'sport', 'Event': 'event'})
sport_df.head()
```

```python
nodup_sport_df = pd.DataFrame.drop_duplicates(sport_df)
nodup_sport_df.head()
```

```python
medals_df = full_athlete[["ID", "Year", "Medal", "Event"]].copy()
medals_df.head()
```

```python
medals_df = medals_df.rename(columns={'ID':'athlete_id', 'Year': 'year', 'Medal':'medal', 'Event':'event'})
medals_df.head()
```

```python
nodup_medals_df = pd.DataFrame.drop_duplicates(medals_df)
nodup_medals_df.head()
```

```python
games_df = full_athlete[["ID", "Year", "NOC", "Games", "Season", "City"]].copy()
games_df.head()
```

```python
games_df = games_df.rename(columns={'ID': 'athlete_id', 'Year':'year', 'NOC':'noc', 'Games':'games', 'Season':'season', 'City':'city'})
games_df.head()
```

```python
nodup_games_df = pd.DataFrame.drop_duplicates(games_df)
nodup_games_df.head()
```

```python
rds_connection_string = "postgres:Jacky0509@localhost:5432/Olympic_DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
regions.to_sql(name='regions', con=engine, if_exists='append', index=False)
```

```python
nodup_teams_df.to_sql(name='teams', con=engine, if_exists='append', index=False)
```

```python
nodup_athlete_df.to_sql(name='athlete', con=engine, if_exists='append', index=False)
```

```python
nodup_sport_df.to_sql(name='sport', con=engine, if_exists='append', index=False)
```

```python
nodup_medals_df.to_sql(name='medals', con=engine, if_exists='append', index=False)
```

```python
nodup_games_df.to_sql(name='games', con=engine, if_exists='append', index=False)
```

```python
nodup_athlete_data.to_sql(name='athlete_data', con=engine, if_exists='append', index=False)
```
