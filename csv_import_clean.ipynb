```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file1 = "Resources/athlete_events.csv"
full_athlete = pd.read_csv(csv_file1)
full_athlete.head()
```

```python
csv_file2 = "Resources/noc_regions.csv"
regions = pd.read_csv(csv_file2)
regions.head()
```

```python
regions = regions.rename(columns={'NOC':'noc'})
regions.head()
```

```python
teams_df = full_athlete[["ID", "Team", "NOC"]].copy()
teams_df.head()
```

```python
teams_df = teams_df.rename(columns={'ID':'athlete_id', 'Team':'team', 'NOC': 'noc'})
teams_df.head()
```

```python
athletes_df = full_athlete[["ID", "Name", "Sex", "Age", "Height", "Weight"]].copy()
athletes_df.head()
```

```python
athletes_df = athletes_df.rename(columns={'ID':'athlete_id', 'Name':'name', 'Sex':'sex','Age':'age', 'Height':'height', 'Weight':'weight'})
athletes_df.head()
```

```python
sport_df = full_athlete[["ID", "Sport"]].copy()
sport_df.head()
```

```python
sport_df = sport_df.rename(columns={'ID':'athlete_id', 'Sport':'sport'})
sport_df.head()
```

```python
medals_df = full_athlete[["ID", "Medal", "Event"]].copy()
medals_df.head()
```

```python
medals_df = medals_df.rename(columns={'ID':'athlete_id', 'Medal':'medal', 'Event':'event'})
medals_df.head()
```

```python
event_df = full_athlete[["ID", "Event"]].copy()
event_df.head()
```

```python
event_df = event_df.rename(columns={'ID':'athlete_id', 'Event':'event'})
event_df.head()
```

```python
games_df = full_athlete[["NOC", "Games", "Year", "Season", "City"]].copy()
games_df.head()
```

```python
games_df = games_df.rename(columns={'NOC':'noc', 'Games':'games', 'Year':'year', 'Season':'season', 'City':'city'})
games_df.head()
```

```python
rds_connection_string = "postgres:Jacky0509@localhost:5432/Olympic_DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
regions.to_sql(name='regions', con=engine, if_exists='append', index=False)
```

```python
teams_df.to_sql(name='teams', con=engine, if_exists='append', index=False)
```

```python
athletes_df.to_sql(name='athletes', con=engine, if_exists='append', index=False)
```

```python
sport_df.to_sql(name='sport', con=engine, if_exists='append', index=False)
```

```python
medals_df.to_sql(name='medals', con=engine, if_exists='append', index=False)
```

```python
games_df.to_sql(name='games', con=engine, if_exists='append', index=False)
```
